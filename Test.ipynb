```python
# Program 1: White Box Testing - Cyclomatic Complexity Calculator

def calculate_complexity(code):
    decision_keywords = ["if", "elif", "for", "while", "and", "or", "except"]

    complexity = 1  # Default complexity starts at 1

    lines = code.split("\n")

    for line in lines:
        line = line.strip()

        for keyword in decision_keywords:
            if line.startswith(keyword):
                complexity += 1

    return complexity


# Example function as input
sample_code = """
def check_number(n):
    if n > 0:
        print("Positive")
    elif n == 0:
        print("Zero")
    else:
        print("Negative")

    for i in range(3):
        print(i)
"""

# Display code
print("Given Function:")
print(sample_code)

# Calculate complexity
cc = calculate_complexity(sample_code)

# Output result
print("McCabe's Cyclomatic Complexity:", cc)
```

```python
# Write a program in Python that performs Black Box Testing by applying
# equivalence partitioning and boundary value analysis.

def validate_age(age):
    """
    Function under test (Black Box)
    Valid age range: 18 to 60
    """
    if 18 <= age <= 60:
        return "Valid Age"
    else:
        return "Invalid Age"


# ---------- Black Box Test Cases ----------

def black_box_testing():
    test_cases = [
        # Equivalence Partitioning
        25,   # Valid partition
        10,   # Invalid partition (too low)
        70,   # Invalid partition (too high)

        # Boundary Value Analysis
        17,   # Lower boundary - 1
        18,   # Lower boundary
        60,   # Upper boundary
        61    # Upper boundary + 1
    ]

    for age in test_cases:
        result = validate_age(age)
        print(f"Age: {age} -> {result}")


# Run tests
black_box_testing()
```

```python
# PROGRAM NO: 3 Automated Software Testing: Generate Test Cases and Validate Output

def add(a, b):
    return a + b


# Automated test simulator
def run_tests():
    test_cases = [
        ((2, 3), 5),
        ((-1, 1), 0),
        ((0, 0), 0),
        ((10, 5), 15)
    ]

    for i, (inputs, expected) in enumerate(test_cases, start=1):
        result = add(*inputs)
        status = "PASS" if result == expected else "FAIL"

        print(
            f"Test {i}: Input={inputs}, Expected={expected}, Actual={result} -> {status}"
        )


run_tests()
```

```python
# Read Test Case from File, Execute It, and Log Result

def add(a, b):
    return a + b

# Create List.txt (only for first run)
with open("List.txt", "w") as f:
    f.write("2,3,5")

# Read test case from file
with open("List.txt", "r") as infile:
    a, b, expected = map(int, infile.readline().split(","))

# Execute test
actual = add(a, b)
result = "PASS" if actual == expected else "FAIL"

# Log result
with open("test_log.txt", "w") as logfile:
    logfile.write(f"Input: {a},{b}\n")
    logfile.write(f"Expected: {expected}\n")
    logfile.write(f"Actual: {actual}\n")
    logfile.write(f"Result: {result}\n")

print("✅ Test executed and logged.")
```

```python
# Program 6: Simple SQA Checklist Program

def check_code(code):
    print("SQA Checklist:\n")

    # Check documentation
    if '"""' in code or "'''" in code:
        print("Documentation: OK")
    else:
        print("Documentation: Missing")

    # Check function naming
    if "def myFunction" in code:
        print("Function Naming: Not proper")
    else:
        print("Function Naming: OK")

    # Check indentation (simple check)
    if "    " in code:
        print("Indentation: Check spacing")
    else:
        print("Indentation: OK")


# Example code
sample_code = '''
def myFunction():
    print("Hello")
'''

check_code(sample_code)
```

```python
# Program 7: Implement McCall's Quality Model
# by evaluating reliability, maintainability and usability

# Input scores out of 10
reliability = float(input("Enter reliability score: "))
maintainability = float(input("Enter maintainability score: "))
usability = float(input("Enter usability score: "))

# Calculate overall quality
overall_quality = (
    reliability + maintainability + usability
) / 3

# Display results
print("\n--- Quality Evaluation ---")
print("Reliability:", reliability)
print("Maintainability:", maintainability)
print("Usability:", usability)
print("Overall Quality Score:", round(overall_quality, 2))
```

```python
# Program 8: Store defects categorized by severity

defects = {
    "critical": [],
    "major": [],
    "minor": []
}


# Function to add a defect
def add_defect(defect_id, description, severity):
    defect = (defect_id, description)

    if severity in defects:
        defects[severity].append(defect)
    else:
        print("Invalid severity level!")


# Function to display defects
def show_defects():
    for severity, defect_list in defects.items():
        print("\n", severity.upper(), "DEFECTS:")

        for defect in defect_list:
            print(
                f"ID: {defect[0]}, "
                f"Description: {defect[1]}"
            )


# Adding defects
add_defect(1, "System crash on login", "critical")
add_defect(2, "UI alignment issue", "minor")
add_defect(3, "Slow response time", "major")

# Display defects
show_defects()
```

```python
# Program 9: Unit testing by adding two numbers

import unittest


# Function to test
def add(a, b):
    return a + b


# Unit test class
class TestAddFunction(unittest.TestCase):

    def test_add(self):
        self.assertEqual(add(2, 3), 5)
        self.assertEqual(add(0, 0), 0)
        self.assertEqual(add(-1, 1), 0)


# Run tests in Jupyter Notebook or Google Colab
suite = unittest.TestLoader().loadTestsFromTestCase(
    TestAddFunction
)

runner = unittest.TextTestRunner(verbosity=2)
runner.run(suite)
```

```python
# Program 10: Simple Calculator Testing

def calc(a, b, operator):
    if operator == "+":
        return a + b

    elif operator == "-":
        return a - b

    else:
        return None


# Test cases
assert calc(5, 3, "+") == 8
assert calc(5, 3, "-") == 2
assert calc(5, 3, "*") is None

print("Calculator tests passed!")
```
