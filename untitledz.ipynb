```python
# Read two numbers and perform arithmetic ops using match-case

# Input two numbers
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

# Display menu
print("\nSelect an operation:")
print("1. Addition")
print("2. Subtraction")
print("3. Multiplication")
print("4. Division")

# Read user's choice
choice = int(input("Enter your choice (1-4): "))

# Perform operation using match-case
match choice:
    case 1:
        print("Addition =", num1 + num2)
    case 2:
        print("Subtraction =", num1 - num2)
    case 3:
        print("Multiplication =", num1 * num2)
    case 4:
        if num2 != 0:
            print("Division =", num1 / num2)
        else:
            print("Error: Division by zero is not allowed.")
    case _:
        print("Invalid choice. Please select a number between 1 and 4.")
```

```python
# Linear Search Program in Python

# List of elements
arr = [12, 45, 23, 67, 34, 89, 10]

# Displaying the array and their indices
print("Array Elements and their Indices:\n")
for i in range(len(arr)):
    print("Index:", i, "Value:", arr[i])

print()

# Element to be searched
target = int(input("Enter the element to search: "))

# Variable to track if element is found
found = False

# Linear search process
for i in range(len(arr)):
    if arr[i] == target:
        print("Element found at index:", i)
        found = True
        break

# If element not found
if not found:
    print("Element not found in the array")
```

```python
# Binary Search Program in Python

# Sorted list of elements
arr = [5, 10, 15, 20, 25, 30, 35, 40, 45]

# Display array and indices
print("Sorted Array Elements and their Indices:\n")
for i in range(len(arr)):
    print("Index:", i, "Value:", arr[i])

print()

# Element to search
target = int(input("Enter the element to search: "))

# Initialize variables
low = 0
high = len(arr) - 1
found = False

# Binary search process
while low <= high:
    mid = (low + high) // 2

    if arr[mid] == target:
        print("Element found at index:", mid)
        found = True
        break

    elif target > arr[mid]:
        low = mid + 1

    else:
        high = mid - 1

# If element not found
if not found:
    print("Element not found in the array")
```

```python
# Selection Sort

# List of numbers
arr = [299, 1009, 14, 37, 13]

print("Original List:", arr)

# Selection sort algorithm
n = len(arr)

for i in range(n):
    # Assume the current position is the minimum
    min_index = i

    # Find the smallest element in the remaining list
    for j in range(i + 1, n):
        if arr[j] < arr[min_index]:
            min_index = j

    # Display array before swapping
    print("Before Swap:", arr)

    # Swap the elements
    arr[i], arr[min_index] = arr[min_index], arr[i]

    # Display array after swapping
    print("After Swap :", arr)

print("\nSorted List in Ascending Order:", arr)
```

```python
# Insertion Sort Program in Python

# List of numbers
arr = [8888, 3125, 5, 2]

print("Original List:", arr)

# Insertion sort algorithm
n = len(arr)

for i in range(1, n):
    key = arr[i]
    j = i - 1

    # Move elements that are greater than key
    # to one position ahead
    while j >= 0 and arr[j] > key:
        arr[j + 1] = arr[j]
        j = j - 1

    arr[j + 1] = key

print("Sorted List in Ascending Order:", arr)
```

```python
# Matrix Multiplication Program in Python

# Matrix A
A = [
    [1, 2],
    [3, 4]
]

# Matrix B
B = [
    [5, 6],
    [7, 8]
]

# Result matrix initialized with zeros
result = [
    [0, 0],
    [0, 0]
]

# Matrix multiplication using nested loops
for i in range(len(A)):
    for j in range(len(B[0])):
        for k in range(len(B)):
            result[i][j] += A[i][k] * B[k][j]

print("Matrix A:")
for row in A:
    print(row)

print("\nMatrix B:")
for row in B:
    print(row)

print("\nResultant Matrix after Multiplication:")
for row in result:
    print(row)
```

```python
# Word Frequency Program in Python

# Create a string
text = "data science is fun and data science is powerful"

# Split the string into words
words = text.split()

# Dictionary to store word frequencies
frequency = {}

# Count the frequency of each word
for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1

# Display the frequency of each word
print("Word Frequencies:")
for word, count in frequency.items():
    print(word, ":", count)

# Find the maximum frequency
max_freq = max(frequency.values())

# Display the most frequent word(s)
print("\nMost Frequent Word(s):")
for word, count in frequency.items():
    if count == max_freq:
        print(word, "appears", count, "times")
```

```python
# Elliptical Orbits

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Semi-major and semi-minor axes
a = 5
b = 3

# Generate parameter values
t = np.linspace(0, 2 * np.pi, 200)

# Ellipse coordinates
x = a * np.cos(t)
y = b * np.sin(t)

# Create figure
fig, ax = plt.subplots()

ax.set_xlim(-6, 6)
ax.set_ylim(-4, 4)
ax.set_aspect("equal")
ax.set_xlabel("X-coordinate")
ax.set_ylabel("Y-coordinate")
ax.set_title("Planetary Motion: Elliptical Orbit")
ax.grid(True)

# Plot the complete orbit path
ax.plot(x, y, label="Orbit")

# Sun at the centre
ax.scatter(0, 0, s=200, marker="o", label="Sun")

# Planet point
planet, = ax.plot([], [], "o", markersize=8, label="Planet")

# Update planet position in each frame
def update(frame):
    planet.set_data([x[frame]], [y[frame]])
    return planet,

# Create animation
ani = FuncAnimation(
    fig,
    update,
    frames=len(t),
    interval=50,
    blit=True,
    repeat=True
)

ax.legend()

# Prevent a static figure from appearing
plt.close(fig)

# Display animation in Jupyter Notebook
HTML(ani.to_jshtml())
```

```python
# Bouncing Ball Simulation in Python

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Initial conditions
height = 10
velocity = 0
gravity = -9.8
time_step = 0.05
bounce_factor = 0.8
y = height

# Create figure
fig, ax = plt.subplots()

ball, = ax.plot([], [], 'o', markersize=12)

ax.set_xlim(-1, 1)
ax.set_ylim(0, 12)
ax.set_title("Bouncing Ball Simulation")
ax.set_xlabel("X-axis")
ax.set_ylabel("Height")
ax.grid(True)

# Update function
def update(frame):
    global y, velocity

    # Update velocity and position
    velocity += gravity * time_step
    y += velocity * time_step

    # Bounce when ball hits the ground
    if y <= 0:
        y = 0
        velocity = -velocity * bounce_factor

    # Update ball position
    ball.set_data([0], [y])

    return ball,

# Create animation
ani = FuncAnimation(
    fig,
    update,
    frames=200,
    interval=50,
    blit=True,
    repeat=True
)

plt.close(fig)

# Display animation in Jupyter Notebook
HTML(ani.to_jshtml())
```

```python
# Regular Expression Pattern Matching Program

import re

# Sample text
text = """
Contact us at support@example.com or admin@test.org.
Call us at 9876543210 for assistance.
Python is widely used in data science and machine learning.
"""

# Regex patterns
email_pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}\b'
phone_pattern = r'\b\d{10}\b'
keyword_pattern = r'\bPython\b'

# Find matches
emails = re.findall(email_pattern, text)
phones = re.findall(phone_pattern, text)
keywords = re.findall(keyword_pattern, text)

# Display results
print("Email Addresses Found:")
for email in emails:
    print(email)

print("\nPhone Numbers Found:")
for phone in phones:
    print(phone)

print("\nKeyword Found:")
for word in keywords:
    print(word)
```

```python
# Demonstration of Test Driven Approach using Python unittest

import unittest

# User-defined function
def square(n):
    return n * n

# Test class
class TestSquareFunction(unittest.TestCase):

    def test_positive_number(self):
        self.assertEqual(square(2), 4)

    def test_zero(self):
        self.assertEqual(square(0), 0)

    def test_negative_number(self):
        self.assertEqual(square(-3), 9)

    # Intentional failed test case
    def test_failed_case(self):
        self.assertEqual(square(4), 10)   # Incorrect expected result

# Run the tests
if __name__ == "__main__":
    unittest.main(argv=['first-arg-is-ignored'], exit=False)
```
